```python
import os
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sqlalchemy import create_engine
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
import sklearn.linear_model as lm
from utils import timeit

plt.rcParams['figure.figsize'] = (12, 9)
plt.rcParams['font.size'] = 12

sns.set(style="whitegrid", palette="muted")
%matplotlib inline
```

Let's use the SQLite engine from part 1

```python
DB_URI = "sqlite:////srv/db/taxi_2016_student_small.sqlite"
TABLE_NAME = "taxi"

sql_engine = create_engine(DB_URI)
with timeit():
    print(f"Table {TABLE_NAME} has {sql_engine.execute(f'SELECT COUNT(*) FROM {TABLE_NAME}').first()[0]} rows!")
```

Queury for individual data frames for each month

```python
jan_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-01-01' AND '2016-02-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """



feb_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-02-01' AND '2016-03-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """

mar_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-03-01' AND '2016-04-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """

apr_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-04-01' AND '2016-05-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """

may_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-05-01' AND '2016-06-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """

jun_query = f"""
SELECT * 
           FROM {TABLE_NAME}
            WHERE tpep_pickup_datetime
                BETWEEN '2016-06-01' AND '2016-07-01'
                AND record_id % 100 == 0
            ORDER BY tpep_pickup_datetime
            """
```

Get all the data frames of the months


```python
with timeit(): # This query should take less than 3 second
    jan_df = pd.read_sql_query(jan_query, sql_engine)

with timeit(): # This query should take less than 3 second
    feb_df = pd.read_sql_query(feb_query, sql_engine)
    
with timeit(): # This query should take less than 3 second
    mar_df = pd.read_sql_query(mar_query, sql_engine)
    
with timeit(): # This query should take less than 3 second
    apr_df = pd.read_sql_query(apr_query, sql_engine)

with timeit(): # This query should take less than 3 second
    may_df = pd.read_sql_query(may_query, sql_engine)

with timeit(): # This query should take less than 3 second
    jun_df = pd.read_sql_query(jun_query, sql_engine)

jan_df['tpep_pickup_datetime'] = pd.to_datetime(jan_df['tpep_pickup_datetime'])
jan_df['tpep_dropoff_datetime'] = pd.to_datetime(jan_df['tpep_dropoff_datetime'])    
    
feb_df['tpep_pickup_datetime'] = pd.to_datetime(feb_df['tpep_pickup_datetime'])
feb_df['tpep_dropoff_datetime'] = pd.to_datetime(feb_df['tpep_dropoff_datetime'])

mar_df['tpep_pickup_datetime'] = pd.to_datetime(mar_df['tpep_pickup_datetime'])
mar_df['tpep_dropoff_datetime'] = pd.to_datetime(mar_df['tpep_dropoff_datetime'])

apr_df['tpep_pickup_datetime'] = pd.to_datetime(apr_df['tpep_pickup_datetime'])
apr_df['tpep_dropoff_datetime'] = pd.to_datetime(apr_df['tpep_dropoff_datetime'])

may_df['tpep_pickup_datetime'] = pd.to_datetime(may_df['tpep_pickup_datetime'])
may_df['tpep_dropoff_datetime'] = pd.to_datetime(may_df['tpep_dropoff_datetime'])

jun_df['tpep_pickup_datetime'] = pd.to_datetime(jun_df['tpep_pickup_datetime'])
jun_df['tpep_dropoff_datetime'] = pd.to_datetime(jun_df['tpep_dropoff_datetime'])
```
